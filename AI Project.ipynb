Numeric & Non-numeric Column

```python
# Prints the numeric columns
df_numeric = df.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print ("Numerical Columns: ", numeric_cols, "\n")
```

```python
# Prints the non-numeric columns
df_non_numeric = df.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print("Non-numerical columns: ", non_numeric_cols)
```

Count of Numeric & Non-numeric Columns

```python
# Assuming df is your DataFrame
total_columns = df.shape[1]

# Select numerical columns
df_numeric = df.select_dtypes(include=[np.number])
numeric_columns_count = len(df_numeric.columns)

# Non-numeric columns count
non_numeric_columns_count = total_columns - numeric_columns_count

print(f"Total columns: {total_columns}")
print(f"Numeric columns: {numeric_columns_count}")
print(f"Non-numeric columns: {non_numeric_columns_count}")
```

Check For Missing Values

```python
# Checks if there are any missing values
# All columns have 0 which means there are no missing values
print("Checking the total of missing values in each column...")
print(df.isnull().sum(), "\n")
# Because our dataset does not have any values that are null we do not need to output the rows that are null.
```

Removing Fields That Have Missing Values

```python
import pandas as pd

print("Checking the total of missing values in each column...")
missing_values_count = df.isnull().sum()
print(missing_values_count, "\n")

# Remove columns with any missing values
df_cleaned = df.dropna(axis=1, how='any')

# Check the result
print(f"Columns before removing missing values: {df.shape[1]}")
print(f"Columns after removing missing values: {df_cleaned.shape[1]}")
```
